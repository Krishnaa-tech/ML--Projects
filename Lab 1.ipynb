```python
he = OneHotEncoder()
new_df = pd.DataFrame(he.fit_transform(df[["GearSystem"]]).toarray())
he.categories_
df = df.join(new_df)
df
```

```python
import pickle
```

```python
pickle.dump(he, open("he.pickle", 'wb'))
```

**Observation** - *Appling One hot Encoder in Gear System*

```python
le = LabelEncoder()
df["Fuel_LE"] = le.fit_transform(df["Fuel"])
```

```python
ge = LabelEncoder()
df["DealingType_LE"] = ge.fit_transform(df["DealingType"])
```

```python
te = LabelEncoder()
df['Owner_LE'] = te.fit_transform(df['NoofOwners'])
```

```python
pickle.dump(le, open("fuel_le.pickle", 'wb'))
pickle.dump(ge,open("dealing_le.pickle",'wb'))
pickle.dump(te,open('Owner_le.pickle','wb'))
```

```python
scalar = MinMaxScaler()
df["OdoMeterReading_Scalar"] = scalar.fit_transform(df[["OdoMeterReading"]])
df
```

```python
pickle.dump(scalar,open('odo_scalar','wb'))
```

**Observation** - *Applying MinMaxScalar on OdoMeterReading to make it into transform variables*

```python
stdscalar = StandardScaler()
df["AgeofCar_SS"] = stdscalar.fit_transform(df[["AgeofCar"]])
df
```

```python
pickle.dump(stdscalar,open('std_scalar','wb'))
```

**Observation** - *StandardScaler removes the mean and scales each feature/variable to unit variance which is applied in AgeofCar.*

```python
price_category = []
for i in df['Price']:
    if i > df['Price'].mean():
        price_category.append("Luxurious")
    else:
        price_category.append("Non-Luxurious")
df['price_category'] = price_category
df
```

```python
y = df['Price']
y
```

```python
df
```

```python
X = df[['AgeofCar_SS','Fuel_LE', 'DealingType_LE','Owner_LE']]
X
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y,  test_size=0.20,random_state = 40)
regressor = LinearRegression()
regressor.fit(train_X, train_y)
pickle.dump(regressor, open("regressor_model.pickle", 'wb'))
```

```python
val_X , val_y
```

```python
train_X , train_y
```

```python
y_pred = regressor.predict(val_X) # Predicting the scores
y_pred
```

```python
# Calculating the accuracy of the model
print('Mean Absolute Error:', metrics.mean_absolute_error(val_y, y_pred)) 
print('Mean Squared Error:' ,metrics.mean_squared_error(val_y, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(val_y, y_pred)))
print("R-2:" ,metrics.r2_score(val_y,y_pred))
```

**Observation** - *The Accuracy Score is 38% which is comparitively low accuracy score.*

```python
# Estimating training and test score 
print("Training Score:",regressor.score(train_X,train_y))
print("Test Score:",regressor.score(val_X,val_y))
```
