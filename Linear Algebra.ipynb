## 1. Blurring

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter

# Read the image for blurring
image = cv2.imread('image.jpg')

# Create a figure with three subplots
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Plot the images on the subplots
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
Gaussian = cv2.GaussianBlur(image, (7, 7), 0)
axes[1].imshow(cv2.cvtColor(Gaussian, cv2.COLOR_BGR2RGB))
axes[1].set_title('Gaussian Blurring')
median = cv2.medianBlur(image, 5)
axes[2].imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
axes[2].set_title('Median Blurring')
median = cv2.medianBlur(image, 5)
axes[2].imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
axes[2].set_title('Median Blurring')
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
axes[3].imshow(cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB))
axes[3].set_title('Bilateral Blurring')

# Show the plot
plt.show()
```

### Another Method

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# load the image into system memory
image = cv2.imread('image.jpg', flags=cv2.IMREAD_COLOR)

# Create a figure with three subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# display the image to the screen
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
plt.axis('off')

# Define the sharpening kernel
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])


# Apply the sharpening kernel to the image
image_sharp = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)

# display the sharpened image to the screen
axes[1].imshow(cv2.cvtColor(image_sharp, cv2.COLOR_BGR2RGB))
axes[1].set_title('Sharpened Image')
plt.axis('off')
plt.show()
```

## 3. Edge Detection 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

 
loaded_image = cv2.imread("image.jpg")
loaded_image = cv2.cvtColor(loaded_image,cv2.COLOR_BGR2RGB)
 
gray_image = cv2.cvtColor(loaded_image,cv2.COLOR_BGR2GRAY)
 
edged_image = cv2.Canny(gray_image, threshold1=30, threshold2=100)
 
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
plt.subplot(1,3,1)
axes[0].imshow(loaded_image,cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,3,2)
axes[1].imshow(gray_image,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,3,3)
axes[2].imshow(edged_image,cmap="gray")
plt.axis("off")
plt.title("Canny Edge Detected Image")
plt.show()
```

## Noise Reduction

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
color_image = cv2.imread('image.jpg')
gray_img = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
sliding_window_size_x = 5
sliding_window_size_y = 5
```

```python
mean_filter_kernel = np.ones((sliding_window_size_x,sliding_window_size_y),np.float32)/(sliding_window_size_x*sliding_window_size_y)
filtered_image = cv2.filter2D(gray_img,-1,mean_filter_kernel)
```

```python
plt.subplot(121),plt.imshow(gray_img),plt.title('Original Noisy Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(filtered_image),plt.title('Filtered Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Image to Matrix

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
src = cv2.imread('image.jpg')
img = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8, 5))
plt.axis("off")
plt.imshow(img)

img = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
plt.imsave('image_gray.jpg', img, cmap='gray')
gsrc = cv2.imread('image_gray.jpg', 0)
img_mat = []
for i in range(0, gsrc.shape[0]):
    row = []
    for j in range(0, gsrc.shape[1]):
        pixel = gsrc.item(i, j)
        row.append(pixel)
    img_mat.append(row)
```

```python
img_mat = np.array(img_mat)
img_mat
```

```python
img_mat.shape
```

```python
img_tran_mat = img_mat.T  #Transpose of the matrix
img_tran_mat
```

```python
cv2.imwrite('image_gray_tran.jpg', img_tran_mat)
```

```python
fig = plt.figure(figsize=(20,15))
orig = cv2.imread('image.jpg')
tran = cv2.imread('image_gray_tran.jpg')
ax1 = fig.add_subplot(2,2,1)
ax1.axis("off")
ax1.title.set_text('Original')
ax1.imshow(orig)
ax2 = fig.add_subplot(2,2,2)
ax2.axis("off")
ax2.title.set_text('Transposed')
ax2.imshow(tran)
```
